```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Initialize a random adjacency matrix 32 x 32

A = np.random.choice([0, 1], size=(32,32), p=[2.5/3, 0.5/3])
np.fill_diagonal(A, 1)

```

```python
#Given an adjacency matrix return the graph with fixed positions

def pos_graph(A):
    
    chan_loc = pd.read_csv("brain_topology.csv",sep=';') 
    chan_loc.drop('Index', axis=1,inplace=True)
    chan_loc = chan_loc[0:32]
    pos = {row[0]: (row[1],row[2]) for row in chan_loc.values}
    dat = pd.DataFrame(A, columns = pos.keys())
    dat.set_index(dat.columns, inplace=True)
    G = nx.Graph()
    G.add_nodes_from(pos.keys())
    for n, p in pos.items():
           G.node[n]['pos'] = p

    G = nx.from_numpy_matrix(dat.values)
    G = nx.relabel_nodes(G, dict(enumerate(dat.columns)))   
    nx.draw(G, pos=pos, with_labels=True, font_size=25, node_size=0)
    plt.show()
    return(nx.info(G))
```

```python
#Visualize the graph
pos_graph(A)
```
